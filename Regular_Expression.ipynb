# **Uses of RegEX :**
* Regular expressions are used to search, match, or extract specific patterns from strings, making them highly useful for text processing tasks such as searching for keywords, extracting portions of strings like email IDs or phone numbers, or validating formats.​

### RegEX can be used to:

* Find patterns within a string: For example, extracting an email address from a longer piece of text, or checking if a particular word or symbol (like '@gmail' or '.com') appears in a string.​

* Match the presence or absence of patterns: RegEX helps verify whether certain patterns are present within text, which is especially helpful for input validation (e.g., whether a string contains a valid phone number or email format).​

* Split strings based on patterns: Using RegEX, strings can be divided into multiple parts wherever a specific pattern occurs, such as splitting data entries or parsing CSV-like lines.​

* Phone number and email verification: It is commonly used to ensure that phone numbers and email addresses entered by users match standard formats, helping with data validation in forms and applications.​

* Web scraping: RegEX is used to extract structured information (such as phone numbers, email addresses, or ID numbers) from unstructured web data by matching predefined patterns.​

### These four are highlighted as major use cases in the lecture:

1. Pattern search within strings
2. Phone number verification
3. Email verification
4. Web scraping of required information

## re.search

- The `search` function scans through the given string and looks for the first location where the regular expression pattern matches.
- It returns a match object if a match is found; otherwise, it returns `None`.
- For example, `re.search(r".", a)` tries to find the first character in string `a` that matches the pattern.
- If a match is found, details like the matched value and its position (start and end index) can be extracted from the match object.

## re.findall

- The `findall` function searches the entire string and returns a list of all non-overlapping matches of the pattern in the string.
- If there are multiple matches, all of them are included in the resulting list.
- If no match is found, it returns an empty list.

# Special Sequences in RegEx

- Special sequences do not match for the actual character in the string; instead, they tell the specific location in the search string where the match must occur.
- It makes it easier to write commonly used patterns.


### Special Sequences in RegEx

| Special Sequence | Description |
|------------------|-------------|
| \A   | Matches if the string begins with the given character |
| \b   | Matches if the word begins or ends with the given character. \b(string) will check for beginning of the word and (string)\b for ending of the word. |
| \B   | Opposite of \b: the string should not start or end with the given regex. |
| \d   | Matches any decimal digit, this is equivalent to the set class [0-9] |
| \D   | Matches any non-digit character, equivalent to [^0-9] |
| \s   | Matches any whitespace character. |
| \S   | Matches any non-whitespace character |
| \w   | Matches any alphanumeric character, equivalent to [a-zA-Z0-9_] |
| \W   | Matches any non-alphanumeric character. |
| \Z   | Matches if the string ends with the given regex |

# **Functions in RegEx :**
1. **re.findall() :**
- Return all non-overlapping matches of pattern in string, as a list of strings. The string is scanned left-to-right, and matches are returned in the order found.

2. **re.compile() :**
- Regular expressions are compiled into pattern objects, which have methods for various operations such as searching for matches or performing string substitutions.

3. **re.split() :**
- Split string by the occurrences of a character or a pattern, upon finding that pattern, the remaining characters from the string are returned as part of the resulting list.

4. **re.sub() :**
- The method returns a string where matched occurrences are replaced with the content of replace variable.

5. **re.subn() :**
- subn() is similar to sub() in all ways, except in its way of providing output. It returns a tuple with a count of the total of replacement and the new string rather than just the string.

6. **re.escape() :**
- Returns string with all non-alphanumerics backslashed, this is useful if you want to match an arbitrary literal string that may have regular expression metacharacters in it.

7. **re.search() :**
- This method either returns None (if the pattern doesn’t match), or a re. This method stops after the first match, so this is best suited for testing a regular expression more than extracting data.

## Difference Between re.search() and re.findall()

| Feature      | re.search()                                         | re.findall()                                           |
|--------------|-----------------------------------------------------|--------------------------------------------------------|
| Return Type  | Returns a Match object or None                      | Returns a list of matching strings or tuples           |
| Match Count  | Returns only the first match                        | Returns all non-overlapping matches                    |
| Use Case     | Best for checking if a pattern exists (testing)      | Best for extracting all matching substrings (extracting data) |
| Groups       | Allows access to groups via `.group()` method        | Returns capturing groups as tuples in the result list  |
| Performance  | Stops at the first match (can be faster)            | Scans the entire string for all possible matches       |

- **re.search()** is ideal when you only need to find out whether a pattern exists in the string or want the first match.
- **re.findall()** is useful when you need to retrieve all occurrences of a pattern from the string.

# Phone Number and Email Verification and Web Scrapping Using RegEx

# 1. Phone Number

```python
phn = "222-444-8787"
```

```python
for i in range(2):
    phn = input("Enter phn number : ")
    if re.search("\d{3}-\d{3}-\d{4}",phn):
        print("It is a verified number")
    else:
        print("Incorrect number")
```

```python
print(re.findall("[\w._%]{0,20}@[\w-].[A-Za-z]{2,3}",email))
```

```python
print(len(re.findall("[\w._%]{0,20}@[\w-].[A-Za-z]{2,3}",email)))
```
